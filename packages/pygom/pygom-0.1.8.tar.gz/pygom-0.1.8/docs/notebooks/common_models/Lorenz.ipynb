# Lorenz

{func}`.Lorenz`

The Lorenz attractor {cite}`Lorenz1963` is defined by these equations.

$$\begin{aligned}
\frac{dx}{dt} &= \sigma (y-x) \\
\frac{dy}{dt} &= x (\rho - z) - y \\
\frac{dz}{dt} &= xy - \beta z
\end{aligned}$$

A classic example is given in the following code.

```python
from pygom import common_models

import numpy

import matplotlib.pyplot as plt

t = numpy.linspace(0, 100, 20000)

ode = common_models.Lorenz({'beta':8.0/3.0, 'sigma':10.0, 'rho':28.0})

ode.initial_values = ([1., 1., 1.], t[0])

solution = ode.integrate(t[1::])

```

```python
f = plt.figure()

plt.plot(solution[:,0], solution[:,2]);

plt.show()
```
